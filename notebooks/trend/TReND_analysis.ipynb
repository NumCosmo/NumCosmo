```python
RICH_CUT_list = [5, 10, 15, 20, 30, 40, 50]
MASS_CUT_list = [1e13, 2e13, 5e13, 8e13, 1e14]

analysis = TA(MASS_CUT_list, RICH_CUT_list)
```

```python
calibration = analysis.mcmc_results("with_correction")    
```

```python
calibration_table = analysis.mcmc_results_table(calibration)
```
