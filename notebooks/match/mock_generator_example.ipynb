```python
from numcosmo_py.catalog import MockGenerator
from numcosmo_py import Nc, Ncm
from numcosmo_py.catalog import sky_match

Omega_b = 0.0486
Omega_c = 0.2614
Omega_k = 0.0
H0 = 67.7

cosmo = Nc.HICosmoDEXcdm.new()
cosmo.omega_x2omega_k()
cosmo["Omegab"] = Omega_b
cosmo["Omegac"] = Omega_c
cosmo["Omegak"] = Omega_k
cosmo["H0"] = H0
cosmo["w"] = -1.0
```

```python
mgen = MockGenerator(
    cosmo,
    halo_set_size = 200,
    cluster_set_size = 100,
    ra_interval = (-10, 10),
    dec_interval = (-75, 75),
    z_interval = (0.2, 0.5),
    cluster_mass_interval = (1e14, 1e15) ,
    halo_mass_interval = (1e13, 1e15),
    seed = 74682   
)
```

```python
clusters = mgen.generate_clusters()
clusters
```

```python
halos = mgen.generate_halos(clusters= clusters)

halos
```

```python
halos_galaxies = mgen.generate_galaxies(halos, object_type='halo')

halos_galaxies
```

```python
cluster_galaxies = mgen.generate_galaxies(clusters, object_type='cluster')

cluster_galaxies
```

```python
halo_commom_galaxies = mgen.get_common_galaxies_by_distance(halos, cluster_galaxies, object_type='halo')

halo_commom_galaxies
```

```python
halo_commom_galaxies = mgen.get_common_galaxies_by_distance(clusters, halos_galaxies, object_type='cluster')

halo_commom_galaxies
```
